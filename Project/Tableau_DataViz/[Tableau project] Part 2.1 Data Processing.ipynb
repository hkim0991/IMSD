# 3. Data Cleaning

## 3.1 Missing Values

```python
data.isnull().sum()
```

```python
# variable 'departure_delay_in_mins'
nan_rows = data[data['departure_delay_in_mins'].isnull()] # in order to see other remarable points among other variables
nan_rows

# It seems like many of rows where missing values are have 'Yes' in 'flight_cancelled' variable
# that is why we don't have any info on 'departure_delay_in_mins', 'arrival_delay_in_mins' and 'flight_time_in_mins'
```

```python
data[data['flight_cancelled'] == 'Yes'].count()

# When a flight has cancelled, we don't have any value in 'arrival_delay_in_mins' and 'flight_time_in_mins'
# but in 'departure_delay_in_mins', there are 56 which have value in 'departure_delay_in_mins'
```

```python
data.loc[(data['flight_cancelled'] == 'Yes') 
         & (data['departure_delay_in_mins'].isnull()) 
         & (data['arrival_delay_in_mins'].isnull()) 
         & (data['flight_time_in_mins'].isnull())]

# 2345/2401 flights are cancelled before the scheduled flight time
```

```python
data.loc[(data['flight_cancelled'] == 'Yes') 
         & ~(data['departure_delay_in_mins'].isnull()) 
         & (data['arrival_delay_in_mins'].isnull()) 
         & (data['flight_time_in_mins'].isnull())]

# 56/2041 flights are cancelled after he scheduled filght time (meaning that people waited and the flight has been cancelled)
```

### Since those Nan values are meaningful and there is no other applicable value that we can replace them,
### we have decided to let them as they are.

```python
data.loc[(data['flight_cancelled'] == 'No') 
         & ~(data['departure_delay_in_mins'].isnull()) 
         & (data['arrival_delay_in_mins'].isnull()) 
         & (data['flight_time_in_mins'].isnull())]

# There are still some Nan in 'arrival_delay_in_mins' and 'flight_time_in_mins' when the flight is not cancelled
# And these flights have been just delayed but we don't know if they have been arrived late or on time
# Hence, we are going to remove those rows with Nan values in 'arrival_delay_in_mins' and 'flight_time_in_mins'
```

```python
data.drop(data.loc[(data['flight_cancelled'] == 'No') 
         & ~(data['departure_delay_in_mins'].isnull()) 
         & (data['arrival_delay_in_mins'].isnull()) 
         & (data['flight_time_in_mins'].isnull())].index, inplace=True)
```

## 3.2 Feature conversions

```python
# 3.2.1 Object to float: 'satisfaction'

data['satisfaction'] = data['satisfaction'].astype('float64')  

# there are the rows which have incorrect values such as 4.002.00
```

```python
data['satisfaction'].unique()
# there are rows with incorrect values such as '4.00.5' and '4.00.2.00'

etrange = ['4.00.5', '4.00.2.00']

# check which rows have those values
data.loc[data['satisfaction'].isin(etrange)]
```

```python
# remove those rows
data.drop(data.loc[data['satisfaction'].isin(etrange)].index, inplace=True)

data['satisfaction'].unique()
```

```python
# alter the type of the variable, 'satisfaction'
data['satisfaction'] = data['satisfaction'].astype('float64')

data.describe()
```

```python
# 3.2.2 String to datetime: 'flight_date'

data['flight_date'] = pd.to_datetime(data['flight_date'])
data.drop(['day_of_month'], inplace=True, axis=1) # drop the column that we don't need

data.info()
```

```python
# check the unique value of the flight year

data['flight_date'].dt.year.unique() 
```

## 3.3 Feature creation

```python
# 3.3.1 'total_delay_in_mins'
data['total_delay_in_mins'] = data['departure_delay_in_mins'] + data['arrival_delay_in_mins']
data.total_delay_in_mins.isnull().sum()
```

```python
# 3.3.2 'flight_status'
data['flight_status'] = np.where(data['total_delay_in_mins'] == 0, 'ontime',
        (np.where(data['total_delay_in_mins'] > 0, 'delayed', 
        (np.where((data['total_delay_in_mins'].isnull()) & (data['flight_cancelled'] == 'No'), 'delayed',
        (np.where((data['total_delay_in_mins'].isnull()) & (data['flight_cancelled'] == 'Yes'), 'cancelled', '')))))))
 
data.flight_status.value_counts()
```

```python
data.drop(['flight_cancelled'], inplace=True, axis=1) # drop the column that we don't need
data.tail(10)
```
